```python
cnbc_df=pd.DataFrame()
```

```python
cnbc_df['content']=None
cnbc_df['key_point']=None
cnbc_df['title']=None
```

```python
for j in range(len(df)):
    article = ''
    key_point = ''
    url=df['URL'][j]
    
    try:
        print(j)
        page = urlopen(url)
        soup = BeautifulSoup(page, 'html.parser')
        
        
        ###main content
        content=soup.find('div', {'class':"ArticleBody-articleBody"})
        
        for i in content.findAll('p'):
            article = article + ' ' +  i.text
        
        cnbc_df = cnbc_df.append({'content': article}, ignore_index=True)
        ######key point
        try:
            key=soup.find('div', {'class':"RenderKeyPoints-keyPoints"})
        
            for i in key.findAll('li'):
                key_point = key_point + ' ' +  i.text

            cnbc_df['key_point'][j]=key_point
        except:
            cnbc_df['key_point'][j]=None
            print("No Key point"+' '+str(j))
            
       
        ###title
        title=soup.find('h1',{'class':'ArticleHeader-headline'})
        cnbc_df['title'][j]=title.text
        print(title.text)
    except:
        print(j,url)
        cnbc_df = cnbc_df.append({'content': None}, ignore_index=True)
        cnbc_df['key_point'][j]=None
        cnbc_df['title'][j]=None
        print("Error opening the URL"+' '+str(j))
        
    
    
    
```

```python
cnbc_df.head(5)
```

```python
cnbc_df['URL']=df['URL']
cnbc_df['DOCID']=df['DOCID']
cnbc_df['LABEL']=df['LABEL']
cnbc_df['PUBLISHED_TIME']=df['PUBLISHED_TIME']
cnbc_df['ENTITYID']=df['ENTITYID']
```
