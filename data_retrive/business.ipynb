```python
cnbc_df=pd.DataFrame()
cnbc_df['content']=None
cnbc_df['title']=None
```

```python
for j in range(len(df)):
    article = ''
    key_point = ''
    url=df['URL'][j]

    #print(j,url)
    try:
        print(j)
        page = urlopen(url)
        soup = BeautifulSoup(page, 'html.parser')
        ###main content
        content=soup.find('div', {'class':"story-content",'itemprop':'articleBody'})

        for i in content.findAll('p'):
            article = article + ' ' +  i.text
        #print(article)
        cnbc_df = cnbc_df.append({'content': article}, ignore_index=True)
        ######key point
       
        title=soup.find('h1', {'itemprop':"headline",'class':'entry-title'})

      
        cnbc_df['title'][j]=title.text
        print(title.text)
      
    except:
       
        print("Error opening the URL"+' '+str(j))
        print(j,url)
```
