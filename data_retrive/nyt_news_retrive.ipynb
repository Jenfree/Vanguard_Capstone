```python
import json
import requests
from os import makedirs
from os.path import join, exists
import datetime, time
from datetime import date, timedelta
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#News API
news_api_key = "2d96952a8950451785cac8a1625268d5"
url = ('https://newsapi.org/v2/everything?'
       'q=tariff&'
       'from=2019-10-04&'
       'sortBy=popularity&'
       'apiKey=%s' %(news_api_key))

response = requests.get(url)

print(response.content)
```

```python
dat_json = response.json()
print(dat_json['articles'][0])
dat_lst = []
for _ in dat_json['articles']:
    cont_lst = []
    link = _['url']
    result = requests.get(link)
    soup = BeautifulSoup(result.content)
    content = soup.find_all("p")
    for c in content:
        cont_lst.append(c.text)
    dat_lst.append('////'.join(cont_lst))
    break
```

```python
#NYT API
api_key = "59Z4xaR9EMf0LZsA2IkAZefRutYjAY0G"

web_url_lst = []
snippet_lst = []
lead_paragraph_lst = []
abstract_lst = []
headline_lst = []
keywords_lst = []
pub_date_lst = []
document_type_lst = []
tmp = '18531109'
```

```python
CTR = 0
while CTR < 5000:
    for _ in range(101):
        print(CTR, tmp, datetime.datetime.now())
        url = ('https://api.nytimes.com/svc/search/v2/articlesearch.json?'
               'q=tariff&'
               'page=%s&'
               'begin_date=%s&'
               'sort=oldest&'
               'api-key=%s' %(str(_), str(tmp), api_key))

        response = requests.get(url)
        dat_json = response.json()
        for doc in dat_json['response']['docs']:
            web_url_lst.append(doc['web_url'])
            snippet_lst.append(doc['snippet'])
            lead_paragraph_lst.append(doc['lead_paragraph'])
            abstract_lst.append(doc['abstract'])
            headline_lst.append(doc['headline'])
            keywords_lst.append(doc['keywords'])
            pub_date_lst.append(doc['pub_date'])
            document_type_lst.append(doc['document_type'])
        date_time_obj = datetime.datetime.strptime(doc['pub_date'], '%Y-%m-%dT%H:%M:%S+0000')
        old_tmp = tmp
        tmp = str(date_time_obj.date())
        tmp = tmp.replace('-','')
        if old_tmp == tmp:
            tmp = str(date_time_obj.date()+datetime.timedelta(days=1))
        time.sleep(6)
        CTR += 1
    df = pd.DataFrame({'web_url': web_url_lst, 'snippet': snippet_lst,
                       'lead_paragraph': lead_paragraph_lst, 'abstract': abstract_lst,
                      'headline': headline_lst, 'keywords': keywords_lst, 'pub_date': pub_date_lst,
                      'document_type': document_type_lst})
    df.to_csv('./nyt_save/NYT_save%s.csv' %(str(CTR)), index = False)
```
