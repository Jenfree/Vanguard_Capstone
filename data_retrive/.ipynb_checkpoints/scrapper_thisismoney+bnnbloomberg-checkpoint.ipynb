```python
import json
import requests
from os import makedirs
from os.path import join, exists
import datetime, time
from datetime import date, timedelta
from bs4 import BeautifulSoup
import pandas as pd
import string
printable = set(string.printable)
filter(lambda x: x in printable, s)
```

```python
def parse_thisismoney(url):
    def get_time(input_str):
        loc = input_str.find(', ')+2
        try:
            parsed_date = datetime.datetime.strptime(input_str[loc:].strip(), '%d %b %Y')
        except ValueError:
            parsed_date = datetime.datetime.strptime(input_str[loc:].strip(), '%d %B %Y')
        return parsed_date
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    timestamp = get_time(soup.find_all('span', class_ = 'article-timestamp article-timestamp-published')[0].get_text())
    title = soup.find_all('p', class_='mol-para-with-font')[0].get_text().encode("utf-8", errors="ignore").decode()
    content = '////'.join([x.get_text().replace('\xa0', ' ').encode("utf-8", errors="ignore").decode() for x in soup.find_all('p', class_='mol-para-with-font')[1:]])
    return title, content, timestamp
```

```python
def parse_bnnbloomberg(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    try:
        timestamp = datetime.datetime.strptime(soup.find('div', class_ = "date").get_text().strip(), "%b %d, %Y")
    except ValueError:
        timestamp = datetime.datetime.strptime(soup.find('div', class_ = "date").get_text().strip(), "%B %d, %Y")
    title = soup.find_all('h1')[-1].get_text()
    try:
        content = '////'.join([x.get_text().replace('\xa0', ' ').encode("utf-8", errors="ignore").decode()\
                               for x in soup.find('div', class_ = "article-text").find_all('p')])
    except AttributeError:
        content = ''
    return title, content, timestamp
```

```python
meta = pd.read_csv('./Columbia_Capstone_2019_data.csv')
```

```python
meta.head()
```

```python
title_lst = []
day_lst = []
month_lst = []
year_lst = []
content_lst = []
ADD_FLAG = False
docid_lst = []
for ctr, (docid, url) in enumerate(zip(meta['DOCID'].tolist(), meta['URL'].tolist())):
    if ctr % 100 == 0: print(ctr,"/", len(meta), "Processed:", len(title_lst))
        
    if 'www.thisismoney.co.uk' in str(url):
        try:
            title, content, timestamp = parse_thisismoney(url)
            ADD_FLAG = True
        except:
            continue
    elif 'www.bnnbloomberg.ca' in str(url):
        try:
            title, content, timestamp = parse_bnnbloomberg(url)
            ADD_FLAG = True
        except:
            continue
    
    if ADD_FLAG:
        title_lst.append(title)
        content_lst.append(content)
        day_lst.append(timestamp.day)
        month_lst.append(timestamp.month)
        year_lst.append(timestamp.year)
        docid_lst.append(docid)
        ADD_FLAG = False
df = pd.DataFrame({'docid': docid_lst, "title": title_lst, "content": content_lst, "year": year_lst, 
                  "month": month_lst, "day": day_lst})
df.to_csv('./67.csv')
```

```python
df = df.drop_duplicates()
```
